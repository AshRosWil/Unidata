## Formatting Text

Paragraphs are separated by a blank line.

You are able to specify *italic*, **bold**, ***bolded italic*** and `monospace` text.

## Lists

You can create numbered and unnumbered lists.

Unnumbered:

* One
* Two
* Three

Numbered:

1. A
1. B
1. C

Embedded Lists:

* A
* B
    * 1
    * 2
    * 3
        1. A
        2. B
        3. C

## Code Blocks

You can also create code blocks using 4 spaces.

    define foobar() {
        print("Hello World");
    }

You can also delineate code blocks with the tilde character.  This will let you specify a language, and gives you syntax highlighting.

~~~python
define foobar() {
    print("Hello World");
}
~~~
