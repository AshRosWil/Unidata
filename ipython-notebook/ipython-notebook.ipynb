# What is IPython Notebook?


![image](files/images/ward/ipython_notebook_image1.png)

*The IPython Notebook is a web-based interactive computational environment where you can combine code execution, text, mathematics, plots and rich media into a single document.*

* The IPython Notebook is a series of *Cells*, each of which may be either a documentation block, or a block of Python code.

Documentation is typically `Markdown`-format, but can also render embedded $\LaTeX$ symbols.


# IPython Notebooks enable sharable, reproducible science

## $\LaTeX$ Example

It is possible to embed LaTeX-style tags in IPython Notebooks.  You must enclose the tags with \$'s.


Code | Result
-----|:------:
A^2 | $A^2$
\frac{1}{2} | $\frac{1}{2}$

```python
%matplotlib inline

from numpy import sin,cos,pi,linspace
from pylab import plot

# Play with these number and ignore the rest
R = 10
r = .4
O = 12

t = linspace(-pi,pi,3000)
x = (R+r)*cos(t) - (r+O)*cos(((R+r)/r)*t)
y = (R+r)*sin(t) - (r+O)*sin(((R+r)/r)*t)
plot(x,y)
```
